```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# null values 
df.isnull().sum()
```

```python
# drop unnamed column 
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

Gender Distribution

```python
# plt.figure(figsize=(4,4))
ax = sns.countplot(data =df ,x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# from above chart we conclude that females are more than males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore": 'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot = True)
plt.show()
```

From above chart we conclude that parents education affects children education 

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore": 'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1,annot = True)
plt.title("Relationship between parents and student education score")
plt.show()
```

There is negligible affect on students of their parents marital status

# Distribution of ethnic group

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()


l = ['group A','group B' , 'group C', 'group D', 'group E']
mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mylist)
plt.pie(mylist,labels = l,autopct = "%1.2f%%")
plt.title("disribution of ethnic groups")
plt.show()
```
